```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
te = pd.read_csv("../results/te_elev.csv")
xc = pd.read_csv("../results/xc_elev.csv")
num_df = pd.read_csv("../results/num_px_elev.csv")#.set_index("elev", inplace = True)
```

```python
num_df = num_df.set_index("elev")
te = te.set_index("elev")
xc = xc.set_index("elev")
```

```python
plt.figure(figsize=(16,5))

# Means
ax0 = plt.subplot(141)
# ax0.set_title("Annual Means", size = 14)

ax0.errorbar(te['ann_smlt_mean'],te.index, xerr = te['ann_smlt_std'], color = 'gray', label = 'snowmelt mean', linewidth = 2,marker='o')

ax0.tick_params(axis='both', which='major', labelsize=12)
ax0.set_ylabel("elevation(m)", size = 15)
ax0.set_xlabel("Mean Annual Snowmelt (mm)", size =14, color = 'gray')
ax0.tick_params(axis='x', colors='gray')

ax1 = ax0.twiny()  # instantiate a second axes that shares the same x-axis
ax1.errorbar(te['ann_prcp_mean'],te.index, xerr = te['ann_prcp_std'], color = 'blue', label = 'rainfall mean', linewidth = 2,marker='o')

ax1.set_xlabel("Mean Annual Precip (mm)", size =14, color = 'blue')
ax1.tick_params(axis='x', colors='blue')

# XC plot 
ax2 = plt.subplot(142)

ax2.errorbar(xc['smlt_mean'],xc.index, xerr = xc['smlt_std'], color = 'gray', label = 'snowmelt mean', linewidth = 2,marker='o')

ax2.tick_params(axis='both', which='major', labelsize=12)
ax2.set_ylabel("")
ax2.set_xlabel("Snowmelt - Baseflow Correlation ", size =14, color = 'gray')
ax2.tick_params(axis='x', colors='gray')

ax3 = ax2.twiny()  # instantiate a second axes that shares the same x-axis
ax3.errorbar(xc['prcp_mean'],xc.index, xerr = xc['prcp_std'], color = 'blue', label = 'rainfall mean', linewidth = 2,marker='o')

ax3.set_xlabel("Rainfall - Baseflow Correlation", size =14, color = 'blue')
ax3.tick_params(axis='x', colors='blue')

ax3.set(yticklabels=[])  
ax3.set_ylabel("")  

# TE plot 
ax4 = plt.subplot(143)

ax4.errorbar(te['smlt_mean'],te.index, xerr = te['smlt_std'], color = 'gray', label = 'snowmelt mean', linewidth = 2,marker='o')

ax4.tick_params(axis='both', which='major', labelsize=12)
ax4.set_ylabel("")
ax4.set_xlabel("Snowmelt - Baseflow TE ", size =14, color = 'gray')
ax4.tick_params(axis='x', colors='gray')
# ax4.set_xlim([0,8])

ax5 = ax4.twiny()  # instantiate a second axes that shares the same x-axis
ax5.errorbar(te['prcp_mean'],te.index, xerr = te['prcp_std'], color = 'blue', label = 'rainfall mean', linewidth = 2,marker='o')

ax5.set_xlabel("Rainfall - Baseflow TE", size =14, color = 'blue')
ax5.tick_params(axis='x', colors='blue')
# ax5.set_xlim([0,8])
ax5.set(yticklabels=[])  
ax5.set_ylabel("")  

plt.subplot(144)
num_df.columns = ['Snowmelt', "Precip"]
ax6 = num_df.plot(kind = 'barh', ax = plt.gca(),color = ['gray', 'blue'])
ax6.tick_params(axis='both', which='major', labelsize=12)
ax6.set(yticklabels=[])  
ax6.set_ylabel("")  
ax6.set_xlabel("Number of pixels")  
ax6.set_title("Number of pixels per area bin", size = 14)
ax6.set_xlim([0,31000])
plt.show()

plt.tight_layout()
```
